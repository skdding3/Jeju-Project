```python
# 월별 업종별로 그룹핑하여 평균 구하기

df_type = df.groupby(['YM', 'Type'], as_index=False).mean()
```

```python
# 총 이용건수와 총 재난지원금 이용건수 컬럼의 정규화 과정  총에서 재난을 뺀 평균
## 월별로 각 컬럼의 최대값으로 나누어 줌

col = ['NumofSpent', 'NumofDisSpent']
num_df_type_5 = (df_type[df_type['YM'] == 202005][col]) / (df_type[df_type['YM'] == 202005][col].max())
num_df_type_6 = (df_type[df_type['YM'] == 202006][col]) / (df_type[df_type['YM'] == 202006][col].max())
num_df_type_7 = (df_type[df_type['YM'] == 202007][col]) / (df_type[df_type['YM'] == 202007][col].max())
num_df_type_8 = (df_type[df_type['YM'] == 202008][col]) / (df_type[df_type['YM'] == 202008][col].max())

norm_df_type_5 = df_type[df_type['YM'] == 202005].copy()
norm_df_type_6 = df_type[df_type['YM'] == 202006].copy()
norm_df_type_7 = df_type[df_type['YM'] == 202007].copy()
norm_df_type_8 = df_type[df_type['YM'] == 202008].copy()

norm_df_type_5[col] = num_df_type_5[col]
norm_df_type_6[col] = num_df_type_6[col]
norm_df_type_7[col] = num_df_type_7[col]
norm_df_type_8[col] = num_df_type_8[col]
```

```python
fig = plt.figure(figsize=(20,20))
ax1 = fig.add_subplot(411)
ax1 = sns.barplot(x='Type', y='NumofSpent', data=norm_df_type_5.nlargest(10, 'NumofSpent'))
ax1.set_title('5월 업종별 평균 이용건수 (NumofSpent)')

ax2 = fig.add_subplot(412)
ax2 = sns.barplot(x='Type', y='NumofSpent', data=norm_df_type_6.nlargest(10, 'NumofSpent'))
ax2.set_title('6월 업종별 평균 이용건수 (NumofSpent)')

ax3 = fig.add_subplot(413)
ax3 = sns.barplot(x='Type', y='NumofSpent', data=norm_df_type_7.nlargest(10, 'NumofSpent'))
ax3.set_title('7월 업종별 평균 이용건수 (NumofSpent)')

ax4 = fig.add_subplot(414)
ax4 = sns.barplot(x='Type', y='NumofSpent', data=norm_df_type_8.nlargest(10, 'NumofSpent'))
ax4.set_title('8월 업종별 평균 이용건수 (NumofSpent)')
plt.show()
```

```python
fig = plt.figure(figsize=(20,20))
ax1 = fig.add_subplot(411)
ax1 = sns.barplot(x='Type', y='NumofDisSpent', data=norm_df_type_5.nlargest(10, 'NumofDisSpent'))
ax1.set_title('5월 업종별 평균 지원금 이용건수 (NumofDisSpent)')

ax2 = fig.add_subplot(412)
ax2 = sns.barplot(x='Type', y='NumofDisSpent', data=norm_df_type_6.nlargest(10, 'NumofDisSpent'))
ax2.set_title('6월 업종별 평균 지원금 이용건수 (NumofDisSpent)')

ax3 = fig.add_subplot(413)
ax3 = sns.barplot(x='Type', y='NumofDisSpent', data=norm_df_type_7.nlargest(10, 'NumofDisSpent'))
ax3.set_title('7월 업종별 평균 지원금 이용건수 (NumofDisSpent)')

ax4 = fig.add_subplot(414)
ax4 = sns.barplot(x='Type', y='NumofDisSpent', data=norm_df_type_8.nlargest(10, 'NumofDisSpent'))
ax4.set_title('8월 업종별 평균 지원금 이용건수 (NumofDisSpent)')
plt.show()
```
