```python
from pyproj import Proj, transform, Transformer
```

```python
transformer = Transformer.from_crs("epsg:5178", "epsg:4326")
```

```python
x1, y1 = 961114.519726,1727112.269174
x2, y2 = transformer.transform(x1,y1)
print(x2,y2)
```

```python
x1, y1 = df['POINT_X'],df['POINT_Y']
x84, y84 = transformer.transform(x1,y1)
print(x84,y84)
```

```python
df['x84'] = x84
df['y84'] = y84
df = df.drop(['POINT_Y','POINT_X'], axis = 1)
ck1 = pd.DataFrame(df.shape, index=("row","columns"))
ck1
```
